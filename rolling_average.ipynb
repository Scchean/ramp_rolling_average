## Data Cleaning & Calculation

```python
# Convert timestamp to datetime 
df_transaction["transaction_time"] = pd.to_datetime(df_transaction["transaction_time"])

# Convert transaction_amount to numeric
# (coerce errors to NaN just in case there are non-numeric strings)
df_transaction["transaction_amount"] = pd.to_numeric(df_transaction["transaction_amount"], errors="coerce")

# Aggregate total transaction amount per day
df_daily = (
    df_transaction
    .groupby(df_transaction["transaction_time"].dt.date)["transaction_amount"]
    .sum()
    .reset_index()
    .rename(columns={"transaction_time": "date", "transaction_amount": "total_amount"})
)

# Sort by date
df_daily = df_daily.sort_values("date")

# Calculate 3-day rolling average
df_daily["rolling_3day_avg"] = df_daily["total_amount"].rolling(window=3).mean()
```

## Result

```python
# Print Extract January 31 result
jan31_avg = df_daily.loc[df_daily["date"] == pd.to_datetime("2021-01-31").date(), "rolling_3day_avg"].values[0]

print("Rolling 3-day average total amount as of Jan 31:", round(jan31_avg, 2))
print(df_daily.tail(5))  # optional, to verify the rolling trend
```
